```python
import pandas as pd
df=pd.read_csv("/content/employee_productivity_dataset.csv")
df.head()
```

```python
## Q1. Handle Missing Values (Basic Cleaning)

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values.
```

```python
##ans.
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```

```python
df["Hours_Worked_Per_Week"] = df["Hours_Worked_Per_Week"].fillna(df["Hours_Worked_Per_Week"].median())
```

```python
df["Performance_Score"] = df["Performance_Score"].fillna(df["Performance_Score"].mean())
```

```python

print(df)
```

```python
##Q2. Label Encoding Convert these columns into numeric values using Label Encoding: Gender Department

##ans.
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df["Gender"] = label_encoder.fit_transform(df["Gender"])
df["Department"] = label_encoder.fit_transform(df["Department"])
print(df)
```

```python
##Q3. One-Hot Encoding

Apply One-Hot Encoding on:

Work_Mode

Location
```

```python
##ans.
df = pd.get_dummies(df, columns=["Work_Mode", "Location"])
print(df)
```

```python
## Q4. Normalization (Min-Max Scaling)

Normalize the following columns:

Salary

Hours_Worked_Per_Week
```

```python
##ans.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["Salary", "Hours_Worked_Per_Week"]] = scaler.fit_transform(df[["Salary", "Hours_Worked_Per_Week"]])
print(df)
```

```python
##Q5. Standardization (Scaling)

Apply StandardScaler on:

Age
Projects_Completed

Display the transformed values.
```

```python
##ans.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[["Age", "Projects_Completed"]] = scaler.fit_transform(df[["Age", "Projects_Completed"]])
print(df)
```

```python
##Q9.

Why is scaling important in Python?
```

## ans.

Scaling is the process of bringing numerical features to a similar range so that no feature dominates others because of its larger values.

Importance of Scaling:
Improves model performance by ensuring all features contribute equally.
Speeds up training for many machine learning algorithms.
Prevents bias toward features with large numerical values.
Helps distance-based algorithms (such as KNN and K-Means) calculate distances accurately.
Improves gradient-based algorithms (such as Logistic Regression, Neural Networks, and SVM) by making optimization faster and more stable.
Example:

Suppose a dataset has:

Age: 20–60
Salary: 20,000–1,00,000

Without scaling, the Salary values dominate because they are much larger than Age. After scaling, both features are on a comparable scale, allowing the model to learn from each feature fairly.

In short: Scaling makes machine learning models more accurate, efficient, and reliable, especially for algorithms that depend on distances or gradient optimization.

##q10.
Why do we convert categorical data into numerical form?

##ans.
We convert categorical data into numerical form because most machine learning algorithms cannot work directly with text or categories. They require numerical input to perform mathematical calculations and make predictions.

Reasons:

Machine learning models accept only numerical data.
It enables algorithms to process categorical features.
It improves model training and prediction.
It makes the dataset compatible with libraries like Scikit-learn.

Categorical data is converted into numerical form so that machine learning algorithms can understand, process, and use it effectively for analysis and prediction.
